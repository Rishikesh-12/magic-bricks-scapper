```python
import requests

from bs4 import BeautifulSoup

r=requests.get("https://www.magicbricks.com/property-for-sale/residential-real-estate?proptype=Multistorey-Apartment,Builder-Floor-Apartment,Penthouse,Studio-Apartment,Residential-House,Villa&Locality=Pari-Chowk&cityName=Greater-Noida")

c=r.content

soup=BeautifulSoup(c,"html.parser")

print(soup.prettify())
```

```python
all=soup.find_all("div",{"class":"flex relative clearfix m-srp-card__container"})

# all[1].find("div",{"class":"m-srp-card__info flex__item"})
print(len(all))
count=0
```

```python
for i in range(len(all)):
    print("Price is : ",all[i].find("div",{"class":"m-srp-card__info flex__item"}).find("div",{"class":"m-srp-card__price"}).text)
    try:
        print("Rate is : ",all[i].find("div",{"class":"m-srp-card__info flex__item"}).find("div",{"class":"m-srp-card__area"}).text)
    except AttributeError:
        print("Rate details not avaulable")
        
        
    print("Type : ",all[i].find("div",{"class":"m-srp-card__desc flex__item"}).find("div",{"class":"m-srp-card__heading clearfix"}).find("h3").find("span",{"class":"m-srp-card__title"}).find("span",{"class":"m-srp-card__title__bhk"}).text.strip())
    y=list(str(all[i].find("div",{"class":"m-srp-card__desc flex__item"}).find("div",{"class":"m-srp-card__heading clearfix"}).find("h3").find("span",{"class":"m-srp-card__title"}).text).split())
    dat=' '.join(map(str,y[3::]))
    print("Description : ",dat.title(),end=" ")
    print()
    x=all[i].find("div",{"class":"m-srp-card__desc flex__item"}).find("div",{"class":"m-srp-card__collapse js-collapse"}).find("div",{"class":"m-srp-card__summary js-collapse__content"}).find_all("div",{"class":"m-srp-card__summary__item"})
    for i in range(0,len(x)):
        print(x[i].find("div",{"class":"m-srp-card__summary__title"}).text.strip().title() + " : ",end='')
        print(x[i].find("div",{"class":"m-srp-card__summary__info"}).text.strip().title())
    
    
    print("----------------------------------------------------------------------------------------------------------")
    count+=1
print(count)
```
